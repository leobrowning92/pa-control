#GTK+ 3 pa GUI test
Official site: [gtk.org](gtk.org)

Tutorial used is found at [link](https://python-gtk-3-tutorial.readthedocs.org/en/latest/index.html)

Documentation [link](https://developer.gnome.org/gtk3/stable/)
and more python specific docs can be found at [link](https://lazka.github.io/pgi-docs/) although the python documentation is not comprehensive

```python
from gi.repository import Gtk, GObject

class paTestGUI(Gtk.Window):

    def __init__(self):
        Gtk.Window.__init__(self, title="pa GUI test") #Top level window
        self.set_border_width(10)
        
       
       
        vbox = Gtk.Box(orientation=Gtk.Orientation.VERTICAL, spacing=6)
        self.add(vbox) #vbox added to parent window
        
        name_hbox = Gtk.Box(spacing=6)
        vbox.pack_start(name_hbox, True, True, 0) #ads hbox to vbox
        
        self.filename=Gtk.Label("Filename : ")
        name_hbox.pack_start(self.filename,True,True,0)
        
        self.entry=Gtk.Entry()#initialize entry widget
        self.entry.set_text("Default Text")
        name_hbox.pack_start(self.entry,True,True,0)#adds entry to vbox
        
        self.enter_button=Gtk.Button(label="Collect Data")
        self.enter_button.connect('clicked',self.on_enter_button_clicked)
        name_hbox.pack_start(self.enter_button,True,True,0)
        
        path_hbox=Gtk.Box(spacing=6)
        vbox.pack_start(path_hbox, True, True, 0)
        
        self.filepath=Gtk.Label("Please select a file path")
        path_hbox.add(self.filepath)
        
        self.folder_browser_button=Gtk.Button(label="Select Folder")
        self.folder_browser_button.connect('clicked',self.on_folder_browser_button_clicked)
        path_hbox.add(self.folder_browser_button)
        
    def on_enter_button_clicked(self,widget):
        #prints the text input to terminal
        print("Enter button pressed, input text : "+ self.entry.get_text())
    
    def on_folder_browser_button_clicked(self,widget):
        
        dialog = Gtk.FileChooserDialog("Please choose a folder", self,
            Gtk.FileChooserAction.SELECT_FOLDER,
            (Gtk.STOCK_CANCEL, Gtk.ResponseType.CANCEL,
             "Select", Gtk.ResponseType.OK))
        dialog.set_default_size(800, 400)

        response = dialog.run()
        if response == Gtk.ResponseType.OK:
            self.filepath.set_text("Folder selected: " +dialog.get_filename())
            print("Select clicked")
            print("Folder selected: " + dialog.get_filename())
            
        elif response == Gtk.ResponseType.CANCEL:
            print("Cancel clicked")

        dialog.destroy()
   

        
win = paTestGUI()
win.connect("delete-event", Gtk.main_quit)
win.show_all()
Gtk.main()
```
